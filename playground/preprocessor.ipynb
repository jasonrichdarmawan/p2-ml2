```python
from keras.models import load_model
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np

preprocessor = load_model("../results/preprocessor.h5")

raw = tf.io.read_file("./data/how-to-convert-word-document-into-image-jpeg-png-gif-tiff.png")
img = tf.io.decode_image(contents=raw)

img = preprocessor(img)
plt.imshow(img)

model = load_model("../results/seq_5.h5")
model.trainable = False

images = np.array([img])

np.argmax(model.predict(images), axis=1)
```
